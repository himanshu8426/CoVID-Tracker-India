```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.mohfw.gov.in/'
```

```python
res = requests.get(url)
soup = BeautifulSoup(res.content, "lxml")
print(soup.prettify())
```

```python
my_table = soup.find_all('table', {'class':'table table-striped table-dark'})
my_table = my_table[-1].get_text()
data = my_table.split('\n')
data.remove('')
```

```python
head = []
state = []
conf_ind = []
conf_for = []
discharged = []
dead = []
print('===============================Header================================')
for i in range(3, 8):
#     print(data[i])
    head.append(data[i])

# Preparing list for each column    
print('===============================States================================')
for i in range(13, len(data), 8):
#     print(data[i])
    state.append(data[i])
state = state[: len(state) - 2] 
state.append('Total number of confirmed cases in India')

print('===========================Confirmed Indian==========================')
for i in range(14, len(data), 8):
#     print(data[i])
    conf_ind.append(data[i])
conf_ind = conf_ind[: len(conf_ind) - 2]

print('==========================Confirmed Foreign==========================')    
for i in range(15, len(data), 8):
#     print(data[i])
    conf_for.append(data[i])
conf_for = conf_for[: len(conf_for) - 2]
    
print('========================Discharged or Migrated=======================')    
for i in range(16, len(data), 8):
#     print(data[i])
    discharged.append(data[i])
discharged = discharged[: len(discharged) - 2]
    
print('================================Death================================')
for i in range(17, len(data), 8):
#     print(data[i])
    dead.append(data[i])
dead = dead[: len(dead) - 2]
```

```python
# Sum of each column
for i in range(0, len(conf_ind)): 
    conf_ind[i] = int(conf_ind[i]) 
total = sum(conf_ind)
conf_ind.append(total)

for i in range(0, len(conf_for)): 
    conf_for[i] = int(conf_for[i]) 
total = sum(conf_for)
conf_for.append(total)

for i in range(0, len(discharged)): 
    discharged[i] = int(discharged[i]) 
total = sum(discharged)
discharged.append(total)

for i in range(0, len(dead)): 
    dead[i] = int(dead[i]) 
total = sum(dead)
dead.append(total)
```

```python
dict = {head[0]: state, 
        head[1]: conf_ind,
        head[2]: conf_for,
        head[3]: discharged,
        head[4]: dead,
        } 
CoVID_df = pd.DataFrame(dict)
CoVID_df
```

```python
out = CoVID_df.to_json(orient='records')
print(out)
with open('COVID Statewise Count.json', 'w+') as f:
    f.write(out)
```
